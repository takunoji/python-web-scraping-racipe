```python
###まとめ

#レシピ一覧のURL
url = 'https://bazurecipe.com/'
#URLをrequest
r = requests.get(url)
time.sleep(3)

#取ってきたHTMLとパーサーを指定
soup = bs4(r.text, 'html.parser')
#soup

#レシピ名称とパスが記載されているclassを取得
contents = soup.find_all(class_="heading heading-secondary")
#contents

recipe_links = []
recipe_titles = []
material_titles = [] #材料とともに表示
amounts = [] ### 分量だけを表示する

for i in range(len(contents)):
    print(str(i)+"回目のループ→")

    get_a = contents[i].find("a")
#    print(get_a)

    #リンクを取得
    for j in range(len(get_a)):
        try:
            link_ = get_a.get("href")
#            print("link_",link_)
            recipe_links.append(link_)
        except:
            pass

    #各レシピのURLから
    #各レシピ名取得_404notFOUND対策ver
#    for k in range(len(recipe_links)):
    recipe_link = recipe_links[i]
    r = requests.get(recipe_link)
    if r == '404 NOT FOUND':
        continue
    else:
        time.sleep(1)
        soup = bs4(r.text, 'html.parser')

        recipe_title = soup.find(class_="heading heading-primary").text
        recipe_titles.append(recipe_title)
        print(recipe_title)
        print(recipe_link)

        #材料を取得する
        #とりあえず1つのURLで
        #url = 'https://erecipe.woman.excite.co.jp/detail/bce7524b3f870ef80a85c6ef79ad1781.html'

#            url = recipe_links[k]
#            print(url)
#            r = requests.get(url)
#            time.sleep(3)
#
#            soup = bs4(r.text, 'html.parser')
        #soup
        for elem in soup.select('div[style^="background"]'):
#                print(elem.attrs, elem.text)

            target = '・'
            idx = elem.text.find(target)  # 半角空白文字のインデックスを検索
            r = elem.text[idx-1:]  # スライスで半角空白文字のインデックス＋1以降を抽出
            # 見つからなかったときを考慮：「r = s[idx+1:] if idx != -1 else 'not found'」
            print(r)

            ### 材料だけを表示する
            sep = ' '
            sep2 = '\n'

            t = r.split(sep)  # 半角空白文字で文字列を分割
            t2 = t[0].split(sep2)  # 半角空白文字で文字列を分割
#                print(t2) 

            target2 = '・'
            target3 = '…'
#                print(len(t2))

            material_title = []                #材料とともに表示
            amount=[]                ### 分量だけを表示する

            target4 = '…'
            target5 = '\n'
#                print(len(t2))
            for l in range(len(t2)):
                s = t2[l]  # スライスで半角空白文字よりも前を抽出
#                    print(s) 
                idx = s.find(target2)
                idx2 = s.find(target3)
                if l!=len(t2):
                    u = s[idx+1:idx2]  # スライスで半角空白文字よりも前を抽出
                else:
                    u = s[idx+2:]  # 最後だけ改行が無い
                print(u) 
                material_title.append(u)

                s = t2[l]  # スライスで半角空白文字よりも前を抽出
                idx = s.find(target4)
                if l!=len(t2):
                    u = s[idx+1:]  # スライスで半角空白文字よりも前を抽出
                else:
                    u = s[idx+2:]  # 最後だけ改行が無い
                print(u) 
                amount.append(u)
            material_titles.append(material_title)
            amounts.append(amount)

            #s2 = r[idx2:idx]  # スライスで半角空白文字よりも前を抽出
            #print(s2)
```

```python
#ここまでをdataFrameに
result ={
    'recipe_title': recipe_titles,
    'recipe_links': recipe_links,
    'material_title': material_titles,
    'amounts': amounts
}
```
