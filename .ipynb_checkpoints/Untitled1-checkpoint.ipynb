```python
from bs4 import BeautifulSoup
#from google.colab import drive
#from google.colab import files
import urllib
import urllib.parse
import urllib.request as req
import csv
import random
import pandas as pd
import numpy as np
import time
import datetime
```

```python
# 調べたい料理名
food_name = 'カレー'
```

```python
url_origin = 'https://recipe.rakuten.co.jp'

url_origin_a = '/?l-id=recipe_list_detail_recipe'

url_origin_b = 'https://recipe.rakuten.co.jp/search/'

url_origin_c = '/?s=4&v=0&t=2'
```

```python
# 各レシピのurlを格納する
recipe_url_lists = []

def GetRecipeURL(url):
    print(url)
    
    res = req.urlopen(url)
    soup = BeautifulSoup(res, 'html.parser')

    # レシピ一覧の範囲を選択
#    recipe_text = str(soup.find_all('li', class_= 'clearfix'))
    recipe_text = str(soup.find_all('li', class_= 'recipe_ranking__item'))
#    print(recipe_text)
    # 取得したテキストを1行ずつ分割してリストに収納
    recipe_text_list = recipe_text.split('\n')
#    print(recipe_text_list)
#    i=0
    # リストを1行ずづ読み込んで料理名が一致する行だけ抽出
    for text in recipe_text_list:
#        i=i+1
#        print(str(i))
#        print(text + str(i))
        # 各レシピのurl取得
        if 'href="/recipe' in text:
            #特定箇所を指定してurlにいれる
            recipe_url_id = text[155:174]
#            print(recipe_url_id)
            # urlの結合
            recipe_url_list = url_origin + recipe_url_id + url_origin_a
#            print(recipe_url_list)
            # urlを格納
            recipe_url_lists.append(recipe_url_list)  

        # 各レシピのタイトル取得
#        if 'h3' in text:
        if 'img alt' in text:
            print(text + ", " + recipe_url_list)
```

```python
# 調べたいページの量
page_count = 2

# 料理名をurlに入れるためエンコード
name_quote = urllib.parse.quote(food_name)

# urlを結合する（1ページのみurl）
# 人気順
base_url = url_origin_b + name_quote
# 新着順
# base_url = 'https://recipe.rakuten.co.jp/search/' + name_quote + '/?s=0&v=0&t=2'

for num in range(page_count):
    #特定のページ以降を取得する場合
    # num = num + 50

    if num == 1:
#        print(base_url)
        # urlを結合する（1ページのみurl）
        GetRecipeURL(base_url)

    if num > 1:
        # urlを結合する（2ページ以降のurl）
        # 人気順
        base_url_other =  url_origin_b + name_quote + '/' + str(num) + url_origin_c
#        print(base_url_other)
        # 新着順
        # base_url_other =  'https://recipe.rakuten.co.jp/search/' + name_quote + '/' + str(num) + '/?s=0&v=0&t=2'
        GetRecipeURL(base_url_other)

        # スクレイピングの1秒ルールを適用
        time.sleep(1)
```

```python
data_count = []
recipe_data_set = []

def SearchRecipeInfo(url, tag, id_name):
    res = req.urlopen(url)
    soup = BeautifulSoup(res, 'html.parser')
    print(url)

    print(tag)
    print(id_name)
    for all_text in soup.find_all(tag, id= id_name):
#        print(all_text)
        
        # ID
#       for text in all_text.find_all('p', class_= 'rcpId'):
        for text in all_text.find_all('li', class_= 'recipe_note__id'):
            recipe_id = text.get_text()[7:17]
            print(recipe_id)

        # 公開日
#       for text in all_text.find_all('p', class_= 'openDate'):
        for text in all_text.find_all('li', class_= 'recipe_note__date'):
            recipe_date = text.get_text()[4:14]
            print(recipe_date)

        # おいしかった, 簡単だった, 節約できたよの3種類のスタンプ数
#       for text in all_text.find_all('div', class_= 'stampHead'):
        for text in all_text.find_all('div', class_= 'recipe_reaction_report'):
#            print(text)
#           for tag in text.find_all('span', class_= 'stampCount'):
            for tag in text.find_all(class_= 'recipe_reaction_button__count is_pushed'):
                if tag.find('span') != None:
                    print(tag.get_text())
                    data_count.append(tag.get_text())

        # つくったよレポート数
#      for text in all_text.find_all('div', class_= 'recipeRepoBox'):
        for text in all_text.find_all('div', class_= 'recipe_info_report__count'):
#           for tag in text.find_all('h2'):
            for tag in text.find_all('a'):
                # レポート数が0のとき
                if tag.find('span') == None:
                    report = str(0)
                else:
                    for el in tag.find('span'):
                        report = el.replace('\n                         ', '').replace('件', '')
    
    print("ID: " + recipe_id + ", DATE: " + recipe_date + ", 作られた数: " + report + 
        ", リピートしたい: " + data_count[0] +
        ", 簡単だった: " + data_count[1] +
        ", 節約できたよ: " + data_count[2]+
        ", url: " + url)
    # csvファイルに書き込むために格納
    recipe_data_set.append([recipe_id, recipe_date, data_count[0], data_count[1], data_count[2], report, url])
    
    # スタンプ数を入れていた配列を空にする
    data_count.clear()

    # スクレイピングの制限
    time.sleep(1)
```

```python
for num in range(len(recipe_url_lists)):
#    SearchRecipeInfo(recipe_url_lists[num], 'div', 'detailContents')
    SearchRecipeInfo(recipe_url_lists[num], 'div', 'recipeDetail')
```
